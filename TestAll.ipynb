```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from Ising1 import IsingGrid
```

```python
ising = IsingGrid(n=256, a=10)
ising.randomflips(0.9)
ising.savegrid('testbefore')
elist = ising.samplegibbs()
ising.savegrid('testafter')
```

```python
plt.subplot()
plt.plot(elist)
plt.autoscale(enable=True, axis='both', tight=None)
plt.show()
```

```python
finalenergylist = []
numbiterlist =[]
for a in np.linspace(0.02, 2, 10):
    ising = IsingGrid(n=256, a=a)
    ising.randomflips(.3)
    elist = ising.samplegibbs()
    numbiterlist.append(len(elist))
    finalenergylist.append(int(elist[-1]/a))
    ising.savegrid('im' + str(int(100*a)).zfill(3))

print(numbiterlist)
print(finalenergylist)
```
